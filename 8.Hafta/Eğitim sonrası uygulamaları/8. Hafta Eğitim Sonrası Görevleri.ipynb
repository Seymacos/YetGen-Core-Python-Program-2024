# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
import datetime
import os

# VKI hesaplama fonksiyonu
def vkiHesapla(kilo, boy_cm):
    boy_m = boy_cm / 100  # Boyu metreye çevir
    return round(kilo / (boy_m * boy_m), 2)

# VKI'ye göre sınıflandırma
def vkiSiniflandirma(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 24.9:
        return "Normal kilolu"
    elif 25 <= vki < 29.9:
        return "Fazla kilolu"
    elif 30 <= vki < 39.9:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

# Tarih formatını kontrol etme fonksiyonu
def tarihKontrol(tarih):
    try:
        datetime.datetime.strptime(tarih, "%d.%m.%Y")
        return True
    except ValueError:
        return False

# Kullanıcıdan veri alma ve VKI hesaplama
def veriAlmaVeIsleme(girisSayisi):
    data = []
    for i in range(girisSayisi):
        isim = input("İsim giriniz: ")
        soyisim = input("Soyisim giriniz: ")
        yas = int(input("Yaşını giriniz: "))
        boy = float(input("Boyunu giriniz (santimetre cinsinden): "))
        kilo = float(input("Kilosunu giriniz: "))
        
        tarih = input("Tarihi giriniz (gün.ay.yıl): ")
        while not tarihKontrol(tarih):
            print("Geçersiz tarih formatı. Lütfen tekrar deneyin.")
            tarih = input("Tarihi giriniz (gün.ay.yıl): ")

        vki = vkiHesapla(kilo, boy)
        sinif = vkiSiniflandirma(vki)
        
        data.append({
            "isim": isim,
            "soyisim": soyisim,
            "yas": yas,
            "boy": boy,
            "kilo": kilo,
            "vki": vki,
            "sinif": sinif,
            "tarih": tarih
        })
        
    # Dosyaya yazma (append mode, mevcut verileri silmeden)
    dosya_mevcut = os.path.exists('vki.txt')
    with open('vki.txt', 'a', encoding='utf-8') as file:
        if not dosya_mevcut:
            file.write(f"{'İsim':<20} {'Soyisim':<20} {'Yaş':<5} {'Boy (cm)':<10} {'Kilo (kg)':<10} {'VKI':<10} {'Sınıf':<20} {'Tarih':<12}\n")
        for entry in data:
            file.write(f"{entry['isim']:<20} {entry['soyisim']:<20} {entry['yas']:<5} {entry['boy']:<10.2f} {entry['kilo']:<10.2f} {entry['vki']:<10.2f} {entry['sinif']:<20} {entry['tarih']:<12}\n")
    
    print("Dosyaya kaydedildi.")

girisSayisi = int(input("Kaç giriş yapacaksınız? "))
veriAlmaVeIsleme(girisSayisi)
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time

menu = {
    "elma": 15,
    "armut": 20,
    "üzüm": 25,
    "çilek": 25,
    "muz": 20,
    "kavun": 30,
    "karpuz": 25
}

# Menü gösterme fonksiyonu
def menuGoster():
    print("\n*Menü*")
    for item in menu:
        print(f"{item}: {menu[item]} TL")

# Satın alma işlemini dosyaya yazdırma fonksiyonu
def dosyayaYaz(urunler, adetler, toplamFiyat):
    with open("kasa.txt", "a", encoding="utf-8") as dosya:
        dosya.write("Satın Alınan Ürünler:\n")
        for urun, adet in zip(urunler, adetler):
            dosya.write(f"{urun} - {adet} adet\n")
        dosya.write(f"Toplam Fiyat: {toplamFiyat} TL\n")
        dosya.write("-" * 40 + "\n")

# Satın alma fonksiyonu
def satinAl():
    menuGoster()
    print("Lütfen satın almak istediğiniz ürünü giriniz. (Çıkış için '3')")

    alinanUrun = []
    alinanAdet = []
    sonFiyat = 0

    while True:
        eklenecekUrun = input("Hangi ürünü eklemek istiyorsunuz? (Çıkış için '3') ")
        if eklenecekUrun == '3':
            break
        if eklenecekUrun in menu:
            try:
                eklemeAdedi = int(input("Kaç adet eklemek istiyorsunuz? "))
                if eklemeAdedi <= 0:
                    print("Pozitif bir sayı giriniz.")
                    continue
            except ValueError:
                print("Lütfen geçerli bir sayı girin.")
                continue
            
            fiyat = menu[eklenecekUrun] * eklemeAdedi
            print(f"Satın alma ücreti: {fiyat} TL.\nSatın almayı onaylamak için 'e', iptal etmek için 'h' giriniz.")
            check = input("Satın almayı onaylıyor musunuz? (e/h) ").strip().lower()
            if check == 'e':
                print("Satın alım onaylandı.")
                print("-" * 40)
                sonFiyat += fiyat
                alinanUrun.append(eklenecekUrun)
                alinanAdet.append(eklemeAdedi)
                print(f"Ara Toplam: {sonFiyat} TL")
            elif check == 'h':
                print("Satın alma işlemi iptal edildi.")
                print("-" * 40)
            else:
                print("Geçersiz giriş, lütfen tekrar deneyin.")
        else:
            print("Ürün menüde bulunmuyor, lütfen geçerli bir ürün girin.")
    
    # Satın alma bilgilerini dosyaya yaz
    dosyayaYaz(alinanUrun, alinanAdet, sonFiyat)
    return sonFiyat, alinanUrun, alinanAdet

def main():
    while True:
        print("\nYetBar'a Hoşgeldiniz")
        print("Menüyü Görüntülemek İçin 1'e Basınız.")
        print("Smoothie Yapmak ve Almak İçin 2'ye Basınız.")
        print("Çıkış Yapmak İçin 3'e Basınız.")
        
        choice = input("Seçiminizi giriniz: ").strip()
        
        if choice == '1':
            menuGoster()

        elif choice == '2':
            toplam_fiyat, urunler, adetler = satinAl()
            print(f"\nToplam Fiyat: {toplam_fiyat} TL")
            print("Satın Alınan Ürünler:")
            for urun, adet in zip(urunler, adetler):
                print(f"{urun} - {adet} adet")

        elif choice == '3':
            print("Çıkış Yapılıyor.")
            break

        else:
            print("Geçersiz seçim, tekrar deneyin.")
            
main()
```
