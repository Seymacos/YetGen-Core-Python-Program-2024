# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
#Vehicle sınıfı
class Vehicle:
    def __init__(self, name, speed, mileage, year, capacity):

        self.name=name
        self.speed=speed
        self.mileage=mileage
        self.year = year
        self.capacity = capacity

#Aracın bilgilerini döndüren method.
    def display_info(self):
        return f"{self.name}: speed: {self.speed} , mileage: {self.mileage} , year: {self.year} , capacity: {self.capacity}"
    
#Vehicle sınıfından inheritance alan bir bus sınıfı
class Bus(Vehicle):
    #capacity 50 değerine varsayılan olarak atandı.
    def __init__(self, name, speed, mileage, year, capacity=50):
        super().__init__(name, speed, mileage, year, capacity)
    
    #Otobüsün kapasitesini döndüren method.
    def seatingCapacity(self):
        return f"The seating capacity of a bus is {self.capacity} passengers.\n"

#Oluşturulan otobüsler.
School_bus = Bus("School Volvo", 180, 12, 2020 )

print(School_bus.display_info())
print(School_bus.seatingCapacity())

School_bus1 = Bus("City Transport", 60, 50000, 2018)

print(School_bus1.display_info())
print(School_bus1.seatingCapacity())
````
