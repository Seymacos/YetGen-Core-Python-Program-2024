# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self, acmaDügmesi="Kapalı", sesAyari=0, kanalListesi=["Kanal D"], kanal="Kanal D"):
        self.acmaDügmesi = acmaDügmesi
        self.sesAyari = sesAyari
        self.kanalListesi = kanalListesi
        self.kanal = kanal

    def tvAc(self):
        if self.acmaDügmesi == "Kapalı":
            print("Televizyon açılıyor.")
            time.sleep(1)
            self.acmaDügmesi = "Açık"
        else:
            print("Televizyon zaten açık durumda.")

    def tvKapat(self):
        if self.acmaDügmesi == "Kapalı":
            print("Televizyon zaten kapalı durumda.")
        else:
            print("Televizyon kapanıyor.")
            time.sleep(1)
            self.acmaDügmesi = "Kapalı"

    def sesAyarlari(self):
        while True:
            sesAyari = input("Sesi azaltmak için '<', sesi yükseltmek için '>' yazınız. İşleminiz bittiyse 'çıkış' yazınız: ")

            if sesAyari == "<":
                if self.sesAyari != 0:
                    self.sesAyari -= 1
                    print("Ses seviyesi:", self.sesAyari)
                else:
                    print("Ses seviyesi 0 olabilecek en düşük seviyede.")

            elif sesAyari == ">":
                if self.sesAyari != 100:
                    self.sesAyari += 1
                    print("Ses seviyesi:", self.sesAyari)
                else:
                    print("Ses seviyesi 100 olabilecek en yüksek seviyede.")

            elif sesAyari == "çıkış":
                print("Ses seviyesi güncellenmiştir. Ses seviyesi:", self.sesAyari)
                break

            else:
                print("Geçersiz işlem.")
            
    def kanalEkle(self, kanalIsmi):
        print("Kanal Ekleniyor...")
        time.sleep(1)
        self.kanalListesi.append(kanalIsmi)
        print("Kanal eklendi!")

    def __len__(self):
        return len(self.kanalListesi)

    def rastgeleKanalaGecme(self):
        rastgele = random.randint(0, len(self.kanalListesi) - 1)
        self.kanal = self.kanalListesi[rastgele]
        print("Rastgele seçilen kanal:", self.kanal)

    def bilgi(self):
        return f"Televizyon {self.acmaDügmesi} durumdadır.\nTelevizyon ses seviyesi {self.sesAyari} değerindedir.\nKanal listesi {self.kanalListesi}\nToplam kanal sayısı {len(self.kanalListesi)}\nŞu anki kanal: {self.kanal}."

kumanda = Kumanda()

print("""
Televizyon Uygulaması
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için 'q' ya basın.
""")

while True:
    islem = input("Lütfen yapmak istediğiniz işlemin numarasını giriniz: ")

    if islem == "q":
        print("Program sonlandırılıyor.")
        break
    elif islem == "1":
        kumanda.tvAc()
    elif islem == "2":
        kumanda.tvKapat()
    elif islem == "3":
        kumanda.sesAyarlari()
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak girin: ")
        kanal_listesi = kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            kumanda.kanalEkle(kanal)
    elif islem == "5":
        print("Kanal sayısı:", len(kumanda))
    elif islem == "6":
        kumanda.rastgeleKanalaGecme()
    elif islem == "7":
        print(kumanda.bilgi())
    else:
        print("Geçersiz işlem.")
```
